```python
#1
class RealNumber:
    
    def __init__(self, r=0):
        self.__realValue = r  
    def getRealValue(self):
        return self.__realValue
    def setRealValue(self, r):
        self.__realValue = r
    def __str__(self):
        return 'RealPart: '+str(self.getRealValue())    
class ComplexNumber(RealNumber):
  def __init__(self, realpart=1.0, imagipart=1.0):
    super().__init__(realpart)
    self.imagipart=imagipart
  def __str__(self):
    return f'Real Part:{float(self.getRealValue())}\nImaginary Part:{float(self.imagipart)}'

cn1 = ComplexNumber()
print(cn1)
print('---------')
cn2 = ComplexNumber(5,7)
print(cn2)

```

```python
#2
class RealNumber:
    def __init__(self, number=0):
        self.number = number
    def __add__(self, anotherRealNumber):
        return self.number + anotherRealNumber.number
    def __sub__(self, anotherRealNumber):
        return self.number - anotherRealNumber.number
    def __str__(self):
        return str(self.number)
class ComplexNumber(RealNumber):
  def __init__(self, realpart, imagipart):
    super().__init__(int(str(realpart)))
    self.realpart=realpart
    self.imagipart=imagipart
  def __str__(self):
    if self.imagipart<0:
      return f'{self.number} - {str(self.imagipart)[1:]}i'
    else:
      return f'{self.number} + {self.imagipart}i'
  def __add__(self, other):
    return ComplexNumber(super().__add__(other), self.imagipart+other.imagipart)
  def __sub__(self, other):
    return ComplexNumber(super().__sub__(other), self.imagipart-other.imagipart)

r1 = RealNumber(3)
r2 = RealNumber(5)
print(r1+r2)
cn1 = ComplexNumber(2, 1)
print(cn1)
cn2 = ComplexNumber(r1, 5)
print(cn2)
cn3 = cn1 + cn2
print(cn3)
cn4 = cn1 - cn2
print(cn4)



```

```python
#5
class Exam:
    def __init__(self,marks):
        self.marks = marks
        self.time = 60
        
    def examSyllabus(self):
        return "Maths , English"
    def examParts(self):
        return "Part 1 - Maths\nPart 2 - English\n"
class ScienceExam(Exam):
  def __init__(self, marks, time, *sub):
    super().__init__(marks)
    self.time=time
    self.sublist= sub
    if len(sub)==2:
      self.sub1= sub[0]
      self.sub2= sub[1]
      self.count= 2+len(sub)
    elif len(sub)==3:
      self.sub1= sub[0]
      self.sub2= sub[1]
      self.sub3= sub[2]
      self.count= 2+len(sub)
  def __str__(self):
    return f'Marks: {self.marks} Time: {self.time} minutes Number of Parts: {self.count}'
  def examParts(self):
    if len(self.sublist)==2:
      return f'{super().examParts()}Part 3 - {self.sub1}\nPart 4 - {self.sub2}'
    elif len(self.sublist)==3:
      return f'{super().examParts()}Part 3 - {self.sub1}\nPart 4 - {self.sub2}\nPart 5 - {self.sub3}'
  def examSyllabus(self):
    extrasub=""
    for i in self.sublist:
      extrasub+=", "+i
    return f'{super().examSyllabus()}, {self.sub1} {extrasub}'
    
    
engineering = ScienceExam(100,90,"Physics","HigherMaths")
print(engineering)
print('----------------------------------')
print(engineering.examSyllabus())
print(engineering.examParts())
print('==================================') 
architecture = ScienceExam(100,120,"Physics","HigherMaths","Drawing")
print(architecture)
print('----------------------------------')
print(architecture.examSyllabus())
print(architecture.examParts())

```

```python
#6
class Shape3D:

  pi = 3.14159
  def __init__(self, name = 'Default', radius = 0):
    self._area = 0
    self._name = name
    self._height = 'No need'
    self._radius = radius   

  def calc_surface_area(self):
    return 2 * Shape3D.pi * self._radius

  def __str__(self):
      return "Radius: "+str(self._radius)
class Sphere(Shape3D):
  def __init__(self, name, radius):
    super().__init__(name,radius)
    self.radius=radius
    self.height= 'No Need'
    print("Shape name: Sphere, Area Formula: 4 * pi * r * r")
  def calc_surface_area(self):
    return super().calc_surface_area()**2/(Shape3D.pi)
  def __str__(self):
    return f'Radius: {self.radius} , Height: {self.height}\nArea: {"{:.3f}".format(self.calc_surface_area())}'

class Cylinder(Shape3D):
  def __init__(self, name, radius, height):
    super().__init__(name, radius)
    self.radius=radius
    self.height=height
    print("Shape Name: Cylinder, Area Formula: 2 * pi * r * (r + h)")
  def __str__(self):
    return f'Radius: {self.radius} , Height: {self.height}\nArea: {"{:.3f}".format(self.calc_surface_area())}'
  def calc_surface_area(self):
    return super().calc_surface_area()*(self.radius+self.height)


sph = Sphere('Sphere', 5)
print('----------------------------------')
sph.calc_surface_area()
print(sph)
print('==================================')
cyl = Cylinder('Cylinder', 5, 10)
print('----------------------------------')
cyl.calc_surface_area()
print(cyl)



```
