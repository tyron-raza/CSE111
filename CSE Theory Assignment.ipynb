```python
class Vector:
  # class variables
  nullx=0.0
  nully=0.0
  nullz=0.0
  __listt=[]

  def __init__(self,x,y,z=0.0):
    self.x = float(x)
    self.y = float(y)
    self.z = float(z)
    Vector.__listt.append((self.x, self.y, self.z))
  
  def is_null(self):
    if self.x == 0 and self.y == 0 and self.z == 0:
      return True
  
  def __str__(self):
    return f'({self.x},{self.y},{self.z})'
  
  def __add__(v1,v2):
    return Vector(v1.x+v2.x, v1.y+v2.y, v1.z+v2.z)
  def __neg__(v1):
    return Vector(-v1.x, -v1.y, -v1.z)
  def __sub__(v1,v2):
    return Vector(v1.x-v2.x, v1.y-v2.y, v1.z-v2.z)
 

  @staticmethod
  def get_null():
    return f'({Vector.nullx},{Vector.nully},{Vector.nullz})'
  
  def from_another_vector(self,a):
    x = self.x + a
    y = self.y + a
    z = self.z + a
    return Vector(x,y,z)
  
  def value(self):
    return self.magnitude()
  
  def magnitude(self):
    magnitude=((self.x)**2+(self.y)**2+(self.z)**2)**.5
    return float(magnitude)
  
  def get_str(self):
    # TODO
    pass

  @staticmethod
  def dot(v1, v2):
    result= v1.x*v2.x + v1.y*v2.y + v1.z*v2.z
    return float(result)
    pass
  
  @staticmethod
  def angle(v1, v2):
    import math
    return math.acos(Vector.dot(v1,v2) / (v1.magnitude() * v2.magnitude())) * 180 / math.pi

  @staticmethod
  def stats():
    print(f'{len(Vector.__listt)} vectors created so far: [')
    for v in Vector.__listt:
      print(v)
      pass
    print(']')
```
