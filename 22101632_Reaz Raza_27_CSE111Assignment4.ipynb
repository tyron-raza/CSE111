```python
class Customer:
  def __init__(self,name):
    self.name=name
  def greet(self, name1=None):
    self.name1=name1
    if self.name1==None:
      print("Hello!")
    else:
      print("Hello", self.name1, "!")
  def purchase(self, *items):
    if len(items)==0:
      print("None")
    else:
      print(self.name, ", you purchased", len(items), "item(s):")
    for i in items:
      print(i)



customer_1 = Customer("Sam")
customer_1.greet()
customer_1.purchase("chips", "chocolate", "orange juice")
print("-----------------------------")
customer_2 = Customer("David")
customer_2.greet("David")
customer_2.purchase("orange juice")
```

```python
class Match:
  def __init__(self, vs):
    print("5..4..3..2..1.. Play !!!")
    self.team1= vs.split("-")[0]
    self.team2= vs.split("-")[1]
    self.runs=1
    self.overs=0
    self.wickets=0

  def add_run(self, add1):
    self.runs+=add1
  def add_over(self, add1):
    if add1==5:
      print("Warning! Cannot add 5 over/s (5 over match)")
    else:
      self.overs+=add1
    
  def add_wicket(self, add1):
    self.wickets+=add1
  def print_scoreboard(self):
    returnval= f'Batting Team:{self.team1} \nBowling Team:{self.team2} \nTotal Runs: {self.runs} Wickets: {self.wickets} Overs: {self.overs}'
    return returnval
      

  

match1 = Match("Rangpur Riders-Cumilla Victorians")
print("=========================")
match1.add_run(4)
match1.add_run(6)
match1.add_over(1)
print(match1.print_scoreboard())
print("=========================")
match1.add_over(5)
print("=========================")
match1.add_wicket(1)
print(match1.print_scoreboard())
```

```python
class Batsman:
  def __init__(self, *info):
    if len(info)==2:
      self.name="New Batsman"
      self.runs=info[0]
      self.balls=info[1]
    else:
      self.name=info[0]
      self.runs=info[1]
      self.balls=info[2]
  def printCareerStatistics(self):
    print(f'Name:{self.name} \nRuns Scored: {self.runs} , Balls Faced: {self.balls}')
  def battingStrikeRate(self):
    return self.runs/self.balls*100
  def setName(self, name1):
    self.name=name1



b1 = Batsman(6101, 7380)
b1.printCareerStatistics()
print("============================")
b2 = Batsman("Liton Das", 678, 773)
b2.printCareerStatistics()
print("----------------------------")
print(b2.battingStrikeRate())
print("============================")
b1.setName("Shakib Al Hasan")
b1.printCareerStatistics()
print("----------------------------")
print(b1.battingStrikeRate())
```

```python
class Account:
  def __init__(self, name= "Default Account", balance= 0.0):
    self.name=name
    self.balance=balance
  def details(self):
    return f'{self.name}\n{self.balance}'
  def withdraw(self, amount):
    if self.balance<=3071 or self.balance-amount==3070
    :
      print("Sorry, Withdraw unsuccessful! The account balance after deducting withdraw amount is equal to or less than minimum.")
    else:
      self.balance-=amount
      print("Withdraw successful! New balance is: ", self.balance)
    


a1 = Account()
print(a1.details())
print("------------------------")
a1.name = "Oliver"
a1.balance = 10000.0
print(a1.details())
print("------------------------")
a2 = Account("Liam")
print(a2.details())
print("------------------------")
a3 = Account("Noah",400)
print(a3.details())
print("------------------------")
a1.withdraw(6930)
print("------------------------")
a2.withdraw(600)
print("------------------------")
a1.withdraw(6929)
```

```python
class StudentDatabase:
  def __init__(self, name, id1):
    self.name=name
    self.id=id1
    self.grades={}
  def calculateGPA(self, *data):
    self.semester= data[1]
    self.courselist=data[0]
    total=0.00
    coursenames=[]
    for i in self.courselist:
      total+=float(i.split(": ")[1])
      coursenames.append(i.split(": ")[0])
    grades0={}
    for i in coursenames:
      grades0[tuple(coursenames)]= "{:.2f}".format(total/len(coursenames))
    for i in data:
      self.grades[data[1]]= grades0
  def printDetails(self):
    print(f'Name:{self.name} \nID: {self.id}')
    for k,v in self.grades.items():
      print("Courses taken in", k, ":")
      for k,v2 in v.items():
        for i in k:
          print(i)
        print("GPA:", v2)
      


s1 = StudentDatabase('Pietro', '10101222')
s1.calculateGPA(['CSE230: 4.0', 'CSE220: 4.0', 'MAT110: 4.0'], 'Summer2020')
s1.calculateGPA(['CSE250: 3.7', 'CSE330: 4.0'], 'Summer2021')
print(f'Grades for {s1.name}\n{s1.grades}')
print('------------------------------------------------------')
s1.printDetails()
s2 = StudentDatabase('Wanda', '10103332')
s2.calculateGPA(['CSE111: 3.7', 'CSE260: 3.7', 'ENG101: 4.0'], 'Summer2022')
print('------------------------------------------------------')
print(f'Grades for {s2.name}\n{s2.grades}')
print('------------------------------------------------------')
s2.printDetails()
```
